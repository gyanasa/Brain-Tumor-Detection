```python
def plotgraph(columns = List[str]):
    fig = make_subplots(rows = 3, cols = 3, subplot_titles = columns)
    row = 1
    col = 1
    for column in columns:
        fig.print_grid
        if col == 4:
            col = 1
            row += 1
        fig.add_trace(go.Scatter(
            x = df[column] ,
            y = df["Chance of Admit "], 
            mode="markers", name = column), 
                      row = row, 
                      col = col)
        col += 1
    fig.update_layout(
        height = 2000, 
        width = 1500, 
        showlegend=False
    )
    fig.print_grid
    fig.show()
    
col = list(df.columns)
col.remove("Serial No.")
col.remove("Chance of Admit ")
plotgraph(col)
```

```python
plt.figure(figsize=(45, 45))
sns.heatmap(df.corr(), annot=True, center=0, linewidths=.5, fmt='.2f', vmin=-1, vmax=1, cmap='RdBu')
```

```python
model = LinearRegression()
y = df.pop("Chance of Admit ")
X = df.drop(["Serial No."], axis=1)
```

```python
predicted = cross_val_predict(model, X, y, cv=10)
```

```python
fig = px.scatter(
    x=y, 
    y=predicted, 
    labels={"x":"Measured", "y":"Predicted"}
)

fig.add_shape(
    type="line", 
    line=dict(dash="dash"),      
    x0 = y.min(), y0 = y.min(),    
    x1 = y.max(), y1 = y.max()
)
fig.update_layout(
    height = 800, 
    width = 800, 
    showlegend=False
)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, 
                                                    test_size = 0.25,
                                                    random_state = 42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
X1 = sm.add_constant(X_train)
result = sm.OLS(y_train, X1).fit()
print(result.summary2())
```

```python
colors = ['Positive' if c > 0 else 'Negative' for c in model.coef_]
fig = px.bar(
    x  = X.columns,
    y = model.coef_,
    labels = dict(x = "Feature", y = "Linear Coefficient", color = colors),
    title = "Weight of each feature for predicting chance of admission")
fig.update_layout(
    height = 800, 
    width = 800, 
    showlegend=False
)
fig.show()
```

```python
compare = pd.DataFrame()
compare["prediction"] = y_pred
compare["measured"] = y_test
fig = px.scatter(
    compare, 
    x = "measured", 
    y = "prediction"
)
fig.update_traces(
    histnorm = "probability", 
    selector = {"type": "histogram"}
)
fig.add_shape(type = "line", 
              line = dict(dash="dash"),
                x0=y.min(), y0=y.min(),
                x1=y.max(), y1=y.max()
)
fig.update_layout(
    height = 800, 
    width = 800, 
    showlegend=False
)
fig.show()
```
